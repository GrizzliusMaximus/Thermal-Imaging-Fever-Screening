```python
path = '8hours/1.txt'
array = np.zeros([288000,8,8])
with open(path) as f:
    for k in range(288000):
        for i in range(8):
            line = f.readline()
            t = [float(x) for x in line.split()]
            array[k,i] = t
        line = f.readline()
#     for line in f:
#         array = [[float(x) for x in line.split()] for line in f]
```
